## Los Grandes Premios (carreras)

```python
races={'season':[],'round':[],'raceName':[],'circuitId':[],'lat':[],'long':[],'country':[],'date':[],'url':[]}
```

```python
warnings.filterwarnings("ignore")
url='http://ergast.com/api/f1/2010.json'
response = requests.get(url)
content=json.loads(response.content)
content['MRData']['RaceTable']['Races']
```

```python
for year in list(range(2010,2023)):
    url='http://ergast.com/api/f1/{}.json'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    for item in content['MRData']['RaceTable']['Races']:
        races['season'].append(int(item['season']))
        races['round'].append(int(item['round']))
        races['raceName'].append(item['raceName'])
        races['circuitId'].append(item['Circuit']['circuitId'])
        races['lat'].append(item['Circuit']['Location']['lat'])
        races['long'].append(item['Circuit']['Location']['long'])
        races['country'].append(item['Circuit']['Location']['country'])
        races['date'].append(item['date'])
        races['url'].append(item['url'])
races
```

## Circuits

```python
circuits={'season':[],'circuitId':[], 
           'name':[],'location':[], 'country':[],
           'lat':[], 'lng':[], 'alt':[],
           'url':[]}
```

```python
warnings.filterwarnings("ignore")
for year in range(2010,20223):
    url= 'http://ergast.com/api/f1/{}/circuits.json?limit=1000'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    path=content['MRData']['CircuitTable']
    for item in range(0,len(path['Circuits'])):
        circuits['season'].append(int(path['season']))
        circuits['circuitId'].append(path['Circuits'][item]['circuitId'])
        circuits['name'].append(path['Circuits'][item]['circuitName'])
        circuits['location'].append(path['Circuits'][item]['Location']['locality'])
        circuits['country'].append(path['Circuits'][item]['Location']['country'])
        circuits['lat'].append(float(path['Circuits'][item]['Location']['lat']))
        circuits['lng'].append(float(path['Circuits'][item]['Location']['long']))
        circuits['alt'].append(path['Circuits'][item]['Location']['locality'])
        circuits['url'].append(path['Circuits'][item]['url'])
```

## Qualifying

```python
qualifiying={"season":[],"driverId":[],"ConstructorId":[],"number":[],"position":[],"qualifyId":[],"raceId":[],"q1":[],"q2":[],"q3":[]}
```

```python
warnings.filterwarnings("ignore")
for year in range(2010,2023):
    url='http://ergast.com/api/f1' 
    url1=url+'/{}.json?limit=1000'
    url3=url+'/{}'.format(year)
    response = requests.get(url1.format(year))
    content=json.loads(response.content)
    tampath=len(content['MRData']['RaceTable']['Races'])
    u=0
    links=[]
    for i in range(1,tampath+1):
        url2=url3+'/{}/qualifying.json?limit=100'
        response = requests.get(url2.format(i))
        content=json.loads(response.content)
        links.append(url2.format(i))
    for j in links:
        response = requests.get(j)
        content=json.loads(response.content)
        for k in content['MRData']['RaceTable']['Races']:
            for k1 in content['MRData']['RaceTable']['Races'][0]['QualifyingResults']:
                qualifiying["season"].append(int(k['season']))
                qualifiying["qualifyId"].append(int( k['round']))
                qualifiying["raceId"].append(k['raceName']) 
                qualifiying["driverId"].append(k1['Driver']['driverId'])   
                qualifiying["ConstructorId"].append(k1['Constructor']['constructorId'])
                qualifiying["number"].append(int(k1['number']))
                qualifiying["position"].append(int(k1['position']))
                qualifiying["q1"].append(k1['Q1'])
                try:
                    qualifiying["q2"].append(k1['Q2'])
                except:
                    qualifiying["q2"].append("")
                try:
                    qualifiying["q3"].append(k1['Q3'])
                except:
                    qualifiying["q3"].append("")
```

```python
for i in qualifiying:
    print(len(qualifiying[i]))
```
