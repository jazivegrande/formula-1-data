## Drivers

```python
drivers ={'season':[],'driverId':[], 'number':[], 'code':[],
          'forename':[], 'surname':[], 'dob':[], 'nationality':[],
          'url':[]}
```

```python
warnings.filterwarnings("ignore")

for year in range(2010,2023):
    url='http://ergast.com/api/f1/{}/drivers.json?limit=1000'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    path=content['MRData']['DriverTable']
    for item in range(0,len(path["Drivers"])):
        drivers["season"].append(int(path['season']))
        drivers['driverId'].append(path['Drivers'][item]["driverId"])
        drivers["forename"].append(path['Drivers'][item]["givenName"])
        drivers["surname"].append(path['Drivers'][item]["familyName"])                       
        drivers["dob"].append(path['Drivers'][item]["dateOfBirth"])                        
        drivers["nationality"].append(path['Drivers'][item]["nationality"])
        drivers["url"].append(path['Drivers'][item]["url"])
        
        try:                   
            drivers['number'].append(int(path['Drivers'][item]['permanentNumber']))
        except:
            drivers["number"].append("")
        try:
            drivers["code"].append(path['Drivers'][item]["code"])
        except:
            drivers["code"].append("")

```

## Qualifying

```python
qualifiying={"season":[],"driverId":[],"ConstructorId":[],"number":[],"position":[],"qualifyId":[],"raceId":[],"q1":[],"q2":[],"q3":[]}
```

```python
warnings.filterwarnings("ignore")
for year in range(2010,2023):
    url='http://ergast.com/api/f1' 
    url1=url+'/{}.json?limit=1000'
    url3=url+'/{}'.format(year)
    response = requests.get(url1.format(year))
    content=json.loads(response.content)
    tampath=len(content['MRData']['RaceTable']['Races'])
    u=0
    links=[]
    for i in range(1,tampath+1):
        url2=url3+'/{}/qualifying.json?limit=100'
        response = requests.get(url2.format(i))
        content=json.loads(response.content)
        links.append(url2.format(i))
    for j in links:
        response = requests.get(j)
        content=json.loads(response.content)
        for k in content['MRData']['RaceTable']['Races']:
            for k1 in content['MRData']['RaceTable']['Races'][0]['QualifyingResults']:
                qualifiying["season"].append(int(k['season']))
                qualifiying["qualifyId"].append(int( k['round']))
                qualifiying["raceId"].append(k['raceName']) 
                qualifiying["driverId"].append(k1['Driver']['driverId'])   
                qualifiying["ConstructorId"].append(k1['Constructor']['constructorId'])
                qualifiying["number"].append(int(k1['number']))
                qualifiying["position"].append(int(k1['position']))
                qualifiying["q1"].append(k1['Q1'])
                try:
                    qualifiying["q2"].append(k1['Q2'])
                except:
                    qualifiying["q2"].append("")
                try:
                    qualifiying["q3"].append(k1['Q3'])
                except:
                    qualifiying["q3"].append("")
```

# Sprint Results

```python
sprintResult={'season':[],#'sprintResultId':[], 
           'raceId':[],'driverId':[], 'constructorId':[],
           'number':[], 'grid':[], 
           'position':[], 'positionText':[],# 'positionOrder':[],
           'points':[],'laps':[], 'time':[], 'milliseconds':[],
           'fastestLap':[],'fastestLapTime':[], 'statusId':[]} 
```

```python
warnings.filterwarnings("ignore")
for year in range(2010,2023):
    url= 'http://ergast.com/api/f1/{}/sprint.json?limit=1000'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    path=content['MRData']['RaceTable']
    for item in range(0,len(path['Races'])):
        for k in path['Races'][item]['SprintResults']:
            sprintResult['season'].append(int(path['season']))
            #sprintResult['sprintResultId'].append(path['Races'][''])
            sprintResult['raceId'].append(path['Races'][item]['raceName'])
            sprintResult['driverId'].append(k['Driver']['driverId'])
            sprintResult['constructorId'].append(k['Constructor']['constructorId'])
            sprintResult['number'].append(int(k['number']))
            sprintResult['grid'].append(k['grid'])
            sprintResult['position'].append(int(k['position']))
            sprintResult['positionText'].append((k['positionText']))
            #sprintResult['positionOrder'].append(k['circuitId'])
            sprintResult['points'].append(int(k['points']))
            sprintResult['laps'].append(int(k['laps']))
            try:
                sprintResult['time'].append(k['Time']['time'])
            except:
                sprintResult["time"].append("")
            try:
                sprintResult['milliseconds'].append(k['Time']['millis'])
            except:
                sprintResult["milliseconds"].append("")
            try:
                sprintResult['fastestLap'].append(int(k['FastestLap']['lap']))
            except:
                sprintResult["fastestLap"].append("")
            try:
                sprintResult['fastestLapTime'].append(k['Time']['time'])
            except:
                sprintResult["fastestLapTime"].append("")
            sprintResult['statusId'].append(k['status'])
```

# Pit Stops

```python
pitStops = {'season':[],"raceId":[],'driverId':[], 'stop':[],
           'lap':[], 'time':[], 'duration':[]}#, 'milliseconds':[]}
```

```python
for year in range(2010,2023):
    url='http://ergast.com/api/f1' 
    url1=url+'/{}.json?limit=1000'
    url3=url+'/{}'.format(year)
    response = requests.get(url1.format(year))
    content=json.loads(response.content)
    tampath=len(content['MRData']['RaceTable']['Races'])
    links=[]
    for i in range(1,tampath+1):
        url2=url3+'/{}/pitstops.json?limit=1000'
        response = requests.get(url2.format(i))
        content=json.loads(response.content)
        links.append(url2.format(i))
    i=0 
    for j in links:
        try:
            response = requests.get(j)
            content=json.loads(response.content)
        except:
            print(f"Hay un error en {j}")
        try: 
            content['MRData']['RaceTable']['Races'][0]['PitStops']
            for item in content['MRData']['RaceTable']['Races'][0]['PitStops']:
                pitStops["raceId"].append(content['MRData']['RaceTable']['Races'][0]['round'])
                pitStops["season"].append(int(content['MRData']['RaceTable']['Races'][0]['season']))
                pitStops["driverId"].append(item['driverId'])
                pitStops["stop"].append(int(item['stop']))
                pitStops["lap"].append(int(item['lap']))
                pitStops["time"].append(item['time'])
                pitStops["duration"].append(item['duration'])
            
        except:
            print(f"Hay un error en {j}")
            
        

pitStopss=pd.DataFrame(pitStops)         
pitStopss.head()
```

# Lap Times

```python
lapTimes = {'season':[],'round':[],'raceId':[],'driverId':[], 'lap':[],
           'position':[], 'time':[]}
```

```python
for year in range(2010,2023):
    url='http://ergast.com/api/f1' 
    url1=url+'/{}.json?limit=1000'
    url3=url+'/{}'.format(year)
    response = requests.get(url1.format(year))
    content=json.loads(response.content)
    links=[]
    for i in range(1,len(content['MRData']['RaceTable']['Races'])+1):
        url2=url3+'/{}/laps.json?limit=10000'
        response = requests.get(url2.format(i))
        content=json.loads(response.content)
        links.append(url2.format(i))
    for j in links:
        response = requests.get(j)
        content=json.loads(response.content)
        for i in content['MRData']['RaceTable']["Races"][0]["Laps"]:
            for item in i['Timings']:
                try:
                    lapTimes["season"].append(int(content['MRData']['RaceTable']['Races'][0]['season']))                    
                except:
                    lapTimes["season"].append("")
                try:
                    lapTimes["round"].append(int(content['MRData']['RaceTable']['Races'][0]['round']))                   
                except:
                    lapTimes["round"].append("")
                try:
                    lapTimes["raceId"].append(content['MRData']['RaceTable']["Races"][0]['raceName'])
                except:
                    lapTimes["raceId"].append("")
                try:
                    lapTimes["driverId"].append(item['driverId'])
                except:
                    lapTimes["driverId"].append("")
                try:
                    lapTimes["lap"].append(int(i['number']))
                except:
                    lapTimes["lap"].append("")
                try:
                    lapTimes["position"].append(int(item['position']))
                except:
                    lapTimes["position"].append("")
                try:
                    lapTimes["time"].append(item['time'])
                except:
                    lapTimes["time"].append("")
                
lapTimes_df=pd.DataFrame.from_dict(lapTimes,orient='index')
```

```python
lapTimes_df=lapTimes_df.transpose()
```

```python
for i in lapTimes_df:
    print(len(lapTimes_df[i]))
```

```python
pd.DataFrame(lapTimes_df).head()
```

```python
pd.DataFrame(lapTimes_df).to_csv('lapTimes.csv',index=False)
```
