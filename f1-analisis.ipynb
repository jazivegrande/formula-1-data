## Los Grandes Premios (carreras)

```python
races={'season':[],'round':[],'raceName':[],'circuitId':[],'lat':[],'long':[],'country':[],'date':[],'url':[]}
```

```python
warnings.filterwarnings("ignore")
url='http://ergast.com/api/f1/2010.json'
response = requests.get(url)
content=json.loads(response.content)
content['MRData']['RaceTable']['Races']
```

```python
for year in list(range(2010,2023)):
    url='http://ergast.com/api/f1/{}.json'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    for item in content['MRData']['RaceTable']['Races']:
        races['season'].append(int(item['season']))
        races['round'].append(int(item['round']))
        races['raceName'].append(item['raceName'])
        races['circuitId'].append(item['Circuit']['circuitId'])
        races['lat'].append(item['Circuit']['Location']['lat'])
        races['long'].append(item['Circuit']['Location']['long'])
        races['country'].append(item['Circuit']['Location']['country'])
        races['date'].append(item['date'])
        races['url'].append(item['url'])
races
```

## Qualifying

```python
qualifying ={'qualifyId':[], 'raceId':[], 'driverId':[],
             'constructorId':[],'number':[],'position':[],
            'q1':[], 'q2':[], 'q3':[]}
```

```python
for year in list(range(2010,2023)):
    url= 'http://ergast.com/api/f1/{}/qualifying.json'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
```

```python
content['MRData']['RaceTable']['Races'][0]['QualifyingResults'][0]['Q1']#['driverId']
```

```python
for item in content['MRData']['RaceTable']['Races']:
        #qualifying['qualifyId'].append(item['qualifyId'])
        qualifying['raceId'].append(item['raceName'])
        qualifying['driverId'].append(item[0]['QualifyingResults'][0]['Driver']['driverId'])
        qualifying['constructorId'].append(item[0]['QualifyingResults'][0]['Constructor']['constructorId'])
        qualifying['number'].append(item[0]['QualifyingResults'][0]['number'])
        qualifying['position'].append(item[0]['QualifyingResults'][0]['position'])
        qualifying['q1'].append(item[0]['QualifyingResults'][0]['Q1'])
        qualifying['q2'].append(item[0]['QualifyingResults'][0]['Q2'])
        qualifying['q3'].append(item[0]['QualifyingResults'][0]['Q3'])
qualifying
```

```python
#qualifying ={'qualifyingId':[],'raceId':[],'driverId':[],'constructorId':[],'number':[],'position':[],'Q1':[],'Q2':[],'Q3':[]}
qualifying ={'driverId':[],'constructorId':[],'number':[],'position':[],'Q1':[],'Q2':[],'Q3':[]}
#qualifying ={'qualifyingId':[],'raceId':[],'driverId':[],'constructorId':[],'number':[],'position':[],'Q1':[],'Q2':[],'Q3':[]}
#qualifying ={'driverId':[],'constructorId':[],'number':[],'position':[],'Q1':[],'Q2':[],'Q3':[]}
qualifying ={'driverId':[],'constructorId':[],'number':[],'position':[],'Q1':[]}
for item in content['MRData']['RaceTable']['Races'][0]['QualifyingResults']:
    #qualifying['qualifyingId'].append(item['qualifyingId'])
    #qualifying['raceId'].append(item['driverRef'])
    qualifying['driverId'].append(item['Driver']['driverId'])
    qualifying['constructorId'].append(item['Constructor']['constructorId'])
    qualifying['number'].append(int(item['number']))
    qualifying['position'].append(int(item['position']))
    qualifying['Q1'].append(item['Q1'])
    #qualifying['Q2'].append(item['Q2'])
    #qualifying['Q3'].append(item['Q3'])
for item in content['MRData']['RaceTable']['Races'][1]['QualifyingResults']:
    #qualifying['qualifyingId'].append(item['qualifyingId'])
    #qualifying['raceId'].append(item['driverRef'])
    qualifying['driverId'].append(item['Driver']['driverId'])
    qualifying['constructorId'].append(item['Constructor']['constructorId'])
    qualifying['number'].append(int(item['number']))
    qualifying['position'].append(int(item['position']))
    qualifying['Q1'].append(item['Q1'])
    #qualifying['Q2'].append(item['Q2'])
    #qualifying['Q3'].append(item['Q3'])
pd.DataFrame(qualifying)
```
